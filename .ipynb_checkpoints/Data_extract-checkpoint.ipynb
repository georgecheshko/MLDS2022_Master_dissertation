```python
# DPD_file
```

```python
years = []
k = 1996
for i in range(24):
    k= k+1
    years.append(k)
```

```python
# c 1996 годом работаем отдельно, т.к там несколько иной вид данных
```

```python
import pandas as pd

# Задать ширину каждого столбца в символах
col_widths = [10, 11, 9, 8, 12, 19, 18]

# Прочитать данные из файла с указанием ширины каждого столбца
dpd_1996 = pd.read_fwf('1996/1996_DPD.txt', widths=col_widths, header=None, skiprows=12)

# Задать заголовки столбцов
dpd_1996.columns = headers = ['Date', '>1 MeV', '>10 MeV', '>100 MeV','>0.6 MeV',
           '>2 MeV', 'Neutron Monitor % of bkgd']
```

```python
result_dpd = dpd_1996
```

```python
for year in years:
    # Задать ширину каждого столбца в символах
    col_widths = [10, 11, 9, 8, 12, 19, 18]
    
    # Прочитать данные из файла с указанием ширины каждого столбца
    dpd = pd.read_fwf(str(year) + '/' + str(year) + '_DPD.txt', widths=col_widths, header=None, skiprows=12)
    
    # Задать заголовки столбцов
    dpd.columns = headers = ['Date', '>1 MeV', '>10 MeV', '>100 MeV','>0.6 MeV',
               '>2 MeV', 'Neutron Monitor % of bkgd']
    result_dpd = pd.concat([result_dpd, dpd])

result_dpd = result_dpd.iloc[:-5]
```

```python
result_dpd
```

```python
# DSD file
```

```python
years = []
k = 1996
for i in range(27):
    k= k+1
    years.append(k)
```

```python
# Задать ширину каждого столбца в символах
col_widths = [10, 5, 8, 10, 6, 8, 8, 3 ,3, 3,3,3,3,3]

# Прочитать данные из файла с указанием ширины каждого столбца
dsd_1996 = pd.read_fwf('1996/1996_DSD.txt', widths=col_widths, header=None, skiprows=13)

# Задать заголовки столбцов
dsd_1996.columns = headers = ['Date', 'Radio_Flux', 'Sunspot_number','Sunspot_area', 'New_Region','Stanford_Solar_Mean_Field',
           'X-ra Flux', 'X-ray_C', 'X-ray_M', 'X-ray_X', 'Optical_S', 'Optical_1', 'Optical_2', 'Optical_3']
result_dsd = dsd_1996
```

```python
for year in years:
    # Задать ширину каждого столбца в символах
    col_widths = [10, 5, 8, 10, 6, 8, 8, 3 ,3, 3,3,3,3,3]
    
    # Прочитать данные из файла с указанием ширины каждого столбца
    dsd = pd.read_fwf(str(year) + '/' + str(year) + '_DSD.txt', widths=col_widths, header=None, skiprows=12)
    
    # Задать заголовки столбцов
    dsd.columns = headers = ['Date', 'Radio_Flux', 'Sunspot_number','Sunspot_area', 'New_Region','Stanford_Solar_Mean_Field',
               'X-ra Flux', 'X-ray_C', 'X-ray_M', 'X-ray_X', 'Optical_S', 'Optical_1', 'Optical_2', 'Optical_3']
    result_dsd = pd.concat([result_dsd, dsd])

result_dsd = result_dsd.iloc[:-5]
```

```python
years = []
k = 1996
for i in range(26):
    k= k+1
    years.append(k)
```

```python
# Задать ширину каждого столбца в символах
col_widths = [10, 8, 2, 2, 2, 2, 2, 2, 2, 2, 6, 3, 2, 2, 2, 2, 2, 2, 2, 6, 3, 2, 2, 2, 2, 2, 2, 2]

# Прочитать данные из файла с указанием ширины каждого столбца
dgd_1996 = pd.read_fwf('1996/1996_DGD.txt', widths=col_widths, header=None, skiprows=12)

# Задать заголовки столбцов
dgd_1996.columns = headers = ['Date', 'ML_A', 'ML_k1','ML_k2','ML_k3','ML_k4','ML_k5','ML_k6','ML_k7','ML_k8','HL_A', 'HL_k1','HL_k2','HL_k3','HL_k4','HL_k5','HL_k6','HL_k7','HL_k8',
                          'E_A', 'E_k1','E_k2','E_k3','E_k4','E_k5','E_k6','E_k7','E_k8']
result_dgd = dgd_1996
```

```python
for year in years:
    # Задать ширину каждого столбца в символах
    col_widths = [10, 6, 3, 2, 2, 2, 2, 2, 2, 2, 6, 3, 2, 2, 2, 2, 2, 2, 2, 6, 3, 2, 2, 2, 2, 2, 2, 2]
    
    # Прочитать данные из файла с указанием ширины каждого столбца
    dgd = pd.read_fwf(str(year) + '/' + str(year) + '_DGD.txt', widths=col_widths, header=None, skiprows=12)
    
    # Задать заголовки столбцов
    dgd.columns = headers = ['Date', 'ML_A', 'ML_k1','ML_k2','ML_k3','ML_k4','ML_k5','ML_k6','ML_k7','ML_k8','HL_A', 'HL_k1','HL_k2','HL_k3','HL_k4','HL_k5','HL_k6','HL_k7','HL_k8',
                              'E_A', 'E_k1','E_k2','E_k3','E_k4','E_k5','E_k6','E_k7','E_k8']
    result_dgd = pd.concat([result_dgd, dgd])

result_dgd = result_dgd.iloc[:-91]
```
